```python
import pandas as pd
df=pd.read_csv("http://13.234.66.67/summer19/datasets/social.csv")
```

```python
df.info()
```

```python
df.head(5)
```

```python
#   removing user id
filterdata=df.iloc[0:,2:].values
```

```python
#   now we can apply training test
features=filterdata[0:,0:2]
```

```python
#   label finding
label=filterdata[0:,2]
```

```python
#   training and testing
from sklearn.model_selection import train_test_split
```

```python
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```python
#   now applt features scalling
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
#    applying scalling in training data only
X=sc.fit_transform(X)
```

```python
X
```

```python
#   testing and scalling
x=sc.transform(x)
```

```python
#   KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#   randon forest call with number of decision tree
rclf=RandomForestClassifier(n_estimators=20)
```

```python
#    now training
trained=rclf.fit(X,Y)
```

```python
#   now predicting
output=trained.predict(x)
```
